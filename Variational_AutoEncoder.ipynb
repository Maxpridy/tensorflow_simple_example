```python
# 

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("./mnist/data/", one_hot=True)
```

```python
learning_rate = 0.01
training_epoch = 20
batch_size = 100


n_hidden = 256
n_z = 20
n_input = 28 * 28
```

```python
def lrelu(x, leak=0.2, name="lrelu"):
    with tf.variable_scope(name):
        f1 = 0.5 * (1 + leak)
        f2 = 0.5 * (1 - leak)
        return f1 * x + f2 * abs(x)
```

```python
def conv2d(x, inputFeatures, outputFeatures, name):
    with tf.variable_scope(name):
        w = tf.get_variable("w",[5,5,inputFeatures, outputFeatures], initializer=tf.truncated_normal_initializer(stddev=0.02))
        b = tf.get_variable("b",[outputFeatures], initializer=tf.constant_initializer(0.0))
        conv = tf.nn.conv2d(x, w, strides=[1,2,2,1], padding="SAME") + b
        return conv
```

```python
def dense(x, inputFeatures, outputFeatures, scope=None, with_w=False):
    with tf.variable_scope(scope or "Linear"):
        matrix = tf.get_variable("Matrix", [inputFeatures, outputFeatures], tf.float32, tf.random_normal_initializer(stddev=0.02))
        bias = tf.get_variable("bias", [outputFeatures], initializer=tf.constant_initializer(0.0))
        if with_w:
            return tf.matmul(x, matrix) + bias, matrix, bias
        else:
            return tf.matmul(x, matrix) + bias
```

```python
def conv_transpose(x, outputShape, name):
    with tf.variable_scope(name):
        # h, w, out, in
        w = tf.get_variable("w",[5,5, outputShape[-1], x.get_shape()[-1]], initializer=tf.truncated_normal_initializer(stddev=0.02))
        b = tf.get_variable("b",[outputShape[-1]], initializer=tf.constant_initializer(0.0))
        convt = tf.nn.conv2d_transpose(x, w, output_shape=outputShape, strides=[1,2,2,1])
        return convt
```

```python
X = tf.placeholder(tf.float32, [None, n_input])
image_matrix = tf.reshape(X, [-1, 28, 28, 1])

with tf.variable_scope("encode"):
    h1 = lrelu(conv2d(image_matrix, 1, 16, "d_h1")) # 28x28x1 -> 14x14x16
    h2 = lrelu(conv2d(h1, 16, 32, "d_h2")) # 14x14x16 -> 7x7x32
    h2_flat = tf.reshape(h2, [batch_size, 7*7*32])

    z_mean = dense(h2_flat, 7*7*32, n_z, "w_mean")
    z_stddev = dense(h2_flat, 7*7*32, n_z, "w_stddev")
    
samples = tf.random_normal([batch_size, n_z], 0, 1, dtype=tf.float32)
guess_z = z_mean + (z_stddev * samples)
 
with tf.variable_scope("decode"):
    z_develop = dense(guess_z, n_z, 7*7*32, scope='z_matrix')
    z_matrix = tf.nn.relu(tf.reshape(z_develop, [batch_size, 7, 7, 32]))
    h1 = tf.nn.relu(conv_transpose(z_matrix, [batch_size, 14, 14, 16], "g_h1"))
    h2 = conv_transpose(h1, [batch_size, 28, 28, 1], "g_h2")
    h2 = tf.nn.sigmoid(h2)
    
generated_images = h2
generated_flat = tf.reshape(generated_images, [batch_size, 28*28])

generation_loss = -tf.reduce_sum(X * tf.log(1e-8 + generated_flat) + (1-X) * tf.log(1e-8 + 1 - generated_flat), 1)

latent_loss = 0.5 * tf.reduce_sum(tf.square(z_mean) + tf.square(z_stddev) - tf.log(tf.square(z_stddev)) - 1, 1)

cost = tf.reduce_mean(generation_loss + latent_loss)
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

total_batch = int(mnist.train.num_examples/batch_size)

for epoch in range(training_epoch):
    total_cost = 0
    
    for i in range(total_batch):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        _, gen_loss, lat_loss = sess.run((optimizer, generation_loss, latent_loss), 
                                         feed_dict={X: batch_xs})
        if i % 100 == 0:
            print("gen_loss :", gen_loss)
            print("lat_loss :", lat_loss)
        
print("done")
```

```python
sample_size = 100

samples = sess.run(generated_images, feed_dict={X: mnist.test.images[:sample_size]})

fig, ax = plt.subplots(2, sample_size, figsize=(sample_size, 2))

for i in range(sample_size):
    ax[0][i].set_axis_off()
    ax[1][i].set_axis_off()
    ax[0][i].imshow(np.reshape(mnist.test.images[i], (28, 28)))
    ax[1][i].imshow(np.reshape(samples[i], (28, 28)))
    
plt.show()
```

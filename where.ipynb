```python
import numpy as np
import random
```

```python
# (32, 65)와 (32, )를 가져와서 만약 (i)가 0이면 (i, 65)를 평탄화함
res = [[random.random() for _ in range(65)] for _ in range(32)]
res = np.array(res)
np.shape(res)

rewards = [0 if i%2==0 else 1 for i in range(32)]
rewards = np.array(rewards)
rewards = np.expand_dims(rewards, axis=1)
print(np.shape(rewards))

result = np.where(np.equal(rewards, 0), res - (res-np.mean(res))/2, res)
for i in result:
    print(i)

#np.where()
#Note: tensorflow and numpy are slightly different
```
